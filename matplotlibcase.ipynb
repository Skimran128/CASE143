```python
import pandas as sk
import matplotlib.pyplot as plt
immu=sk.read_csv('energy.csv')
print(immu)
a=immu[immu['State']=='Andhra']
b=immu[immu['State']=='Karnataka']
c=immu[immu['State']=='Kerala']
d=immu[immu['State']=='Tamilnadu']
fig,diag=plt.subplots(2,2)
#plt.subplots_adjust(wspace=0.8, hspace=0.3)
#plt.subplots_adjust(left=0.2, right=0.8, top=0.9, bottom=0.1)

# Iterate over each axes object and hide the x-axis
for ax in fig.axes:
    ax.get_xaxis().set_visible(False)

diag[0,0].plot(a['Date'],a['Consumption'],color="red")
diag[0,1].plot(b['Date'],b['Consumption'])
diag[1,0].plot(c['Date'],c['Consumption'],color="yellow")
diag[1,1].plot(d['Date'],d['Consumption'],color="green")

fig,diag=plt.subplots(2,2)
for ax in fig.axes:
    ax.get_xaxis().set_visible(False)
diag[0,0].bar(a['Date'],a['Consumption'],color="red")
diag[0,1].bar(b['Date'],b['Consumption'])
diag[1,0].bar(c['Date'],c['Consumption'],color="yellow")
diag[1,1].bar(d['Date'],d['Consumption'],color="green")
#pie
s1='Odisha'
s2='Punjab'
s=(s1,s2)
a=immu[immu['State']=='Andhra']['Consumption'].sum()
b=immu[immu['State']=='Sikkim ']['Consumption'].sum()
s=(a,b)
fig,diag=plt.subplots(2)
diag[0].pie(s,labels=s,autopct='%1.1f%%')
```
